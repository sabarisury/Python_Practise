```python
import matplotlib.pyplot as plt

# First plot
plt.figure()  # New figure
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Plot 1')
plt.show()

# Second plot
plt.figure()  # New figure
plt.plot([1, 2, 3], [7, 8, 9])
plt.title('Plot 2')
plt.show()
```

```python
plt.figure(figsize=(8, 10), dpi=50)  # 8x6 inches, 100 DPI
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
plt.figure()
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Subplot 1')
plt.subplot(1, 2, 2)  # 2nd subplot
plt.plot([1, 2, 3], [7, 8, 9])
plt.title('Subplot 2')
plt.show()
```
