```python
# Step 15. Weekly stats across all locations (first 52 weeks)
weekly_stats = data.resample('W-MON', origin='1961-01-02').agg(['mean', 'max', 'min', 'std'])

# Now take stats across all locations (axis=1)
weekly_stats = weekly_stats.mean(axis=1).to_frame('Mean').join([
    weekly_stats.max(axis=1).to_frame('Max'),
    weekly_stats.min(axis=1).to_frame('Min'),
    weekly_stats.std(axis=1).to_frame('Std')
])

# Keep only first 52 weeks
#weekly_stats = weekly_stats.iloc[:52]

print(weekly_stats.head())
```
