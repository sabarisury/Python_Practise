```python
# Group by Name and sum counts
b = baby_names.groupby("Name")["Count"].sum()
print(b.head())

# Find the minimum count
min_b = b.min()
print("Minimum count:", min_b)

# How many names have this minimum count
num_names = (b == min_b).sum()
print("Number of names with this minimum:", num_names)
```

```python
# Group by name and sum counts
names = baby_names.groupby("Name")[["Count"]].sum()

# Find the median occurrence
median_occurrence = names["Count"].median()

# Get names that have exactly the median occurrence
median_names = names[names["Count"] == median_occurrence]

print("Median occurrence:", median_occurrence)
print("Number of names with this occurrence:", median_names.shape[0])
print(median_names.head())
```
