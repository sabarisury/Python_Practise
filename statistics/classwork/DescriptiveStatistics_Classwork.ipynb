```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="sin(x)")

# Labels & title
plt.title("Sine Wave")
plt.xlabel("X values")
plt.ylabel("Y values")

# Grid & reference lines
plt.grid(True)
plt.axhline(0, color="black", linestyle="--")
plt.axvline(5, color="red", linestyle=":")

# Axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Legend
plt.legend()

plt.show()
```

```python
plt.figure()
plt.hist(df['math score'],bins=50)
plt.title('Histogram graph with maths score')
plt.xlabel("Marks in range bar")
plt.ylabel("Frequecy of marks in specific bar range")
plt.axvline(df["math score"].mean(),color='red',linestyle='--')
plt.axvline(df["math score"].median(),color='black',linestyle='dotted')
plt.axvline(df["math score"].mode()[0],color='yellow')
```
