```python
#series

import pandas as pd

data = [10,20,30,40,50]

s = pd.Series(data)
print(s)

#named series means have column name
sn = pd.Series(data,name="Scores")
print(sn)
```

```python
#dataframe
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Department": ["HR", "IT", "Finance"]
}

df = pd.DataFrame(data)
print(df)
```

```python
#basic exploration
display(df.head())
print("*" * 100)
display(df.columns)
print("*" * 100)
display(df.shape)
print("*" * 100)
display(df.info())
print("*" * 100)
display(df.describe())
```

```python
#modifying columns and rows
#renaming the columns
df = df.rename(columns={"Department" : "dept"}) #columns expect key and value pair
#df = df.rename(columns={"old_name": "new_name", "old_name2": "new_name2"})
df.head()
```

```python
#droping the column -- columns expect list here
#df.drop(columns=["col1", "col2"])   # drop columns
#df.drop(index=[1, 2, 3])           # drop rows by index
df = df.drop(columns=["Age"])
df.head()
```

```python
#filter the rows

it_emp = df[df["dept"]=="IT"]
it_emp_bool = df["dept"]=="IT"

```
