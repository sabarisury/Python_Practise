```python
retail_data["TotalPrice"] = retail_data.apply(lambda row: row["Quantity"] * row["UnitPrice"], axis=1)
```

```python
retail_data["TotalPrice"] = retail_data.apply(lambda x : x["Quantity"] * x["UnitPrice"] ,axis=1)
```

```python
retail_data_f = retail_data[retail_data['TotalPrice']>1000]
retail_data_f.head()
```

```python
retail_data.sort_values(by="TotalPrice",ascending=False).head()
```

```python
retail_data.groupby("Country").agg({"Quantity" : "sum", "TotalPrice" :"sum"}).sort_values(by="TotalPrice",ascending=False).head()
```

```python
retail_data.groupby("CustomerID")["TotalPrice"].sum("TotalPrice").sort_values(by="TotalPrice",ascending=False)
```

```python
retail_data.groupby("CustomerID")["TotalPrice"].sum().sort_values(ascending=False).head()
```

```python
retail_data["InvoiceDate"] = pd.to_datetime(retail_data["InvoiceDate"], format="%d/%m/%Y %H:%M", dayfirst=True)
```

```python
retail_data.info()
```

```python
retail_data["invoice_month"] = retail_data["InvoiceDate"].dt.month.astype("Int64")
```

```python
retail_data.head()
```

```python
retail_data["invoice_month"].value_counts()
```

```python
retail_data.groupby("invoice_month")["InvoiceNo"].count()
```

```python
retail_data.groupby("invoice_month")["InvoiceNo"].nunique()
```

```python
txn_per_month = (
    retail_data.groupby("invoice_month")["InvoiceNo"]
    .nunique()
    .reset_index(name="transaction_count")
)

print(txn_per_month)
```
