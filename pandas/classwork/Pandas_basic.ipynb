**Series in pandas**

```python
import pandas as pd

data = pd.Series([1,2,3,4,5])
print(type(data))
print(data)

#series will have data with default numbered index from zero

data_2 = pd.Series([1,2,3,4,5],index=["a","b","c","d","e"])
print(data_2)

#accesing data
print("accessing data")
print(data[0])
print(data[0:2])

print(data_2["a"])
print(data_2["a":"d"])
```

**DATAFRAME in pandas**

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data,index= ["a","b","c"])
df.head()
```

```python
#Accessing Data in DataFrames

"""
df['column']: Access a column.
df.loc[row_label]: Access by label.
df.iloc[row_index]: Access by integer index.
"""
print(df["Age"])
print("*******")
print(df.loc["a"])
print(df.loc["b"]) ##local
print("********")
print(df.iloc[0]) ## like indexloc
```

```python
df.head()
df.tail()
df.describe(include="all")
```
