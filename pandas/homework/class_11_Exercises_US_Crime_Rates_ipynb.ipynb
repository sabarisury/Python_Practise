```python
import pandas as pd
```

```python
df = pd.read_csv("/content/US_Crime_Rates_1960_2014.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Year'] = pd.to_datetime(df['Year'],format='%Y')
```

```python
df.info()
```

```python
df['Year'].head()
```

```python
df = df.set_index('Year')
```

```python
df.head()
```

```python
df = df.drop("Total",axis=1)
```

```python
df.head()
```

```python
df['Decade'] = df.index.year//10
```

```python
df.head()
```

```python
df.groupby("Decade").sum()
```

```python
crime_cols = [
    "Murder", "Forcible_Rape", "Robbery",
    "Aggravated_assault", "Burglary",
    "Larceny_Theft", "Vehicle_Theft"
]

df_decade = df.groupby("Decade")[crime_cols].sum()
df_decade.head()
```

```python
df_decade["Total_Crimes"] = df_decade.sum(axis=1)
```

```python
df_decade
```

```python
print(df_decade["Total_Crimes"].idxmax())
```

```python
x = df_decade.index.where(df_decade['Total_Crimes'] == df_decade['Total_Crimes'].max())[3]
```
